```python
import os
import numpy as np
import pickle as pkl
from matplotlib import pyplot as plt
import pandas as pd
```

# Data Split: Test

```python
test_male = open('SGNet_JAAD_test_male.pkl', 'rb')
test_male = pkl.load(test_male)
test_female = open('SGNet_JAAD_test_female.pkl', 'rb')
test_female = pkl.load(test_female)
```

## Statistics

#### Number of trajectories

```python
print('Male:', str(len(test_male['MSE']['ADE_05'])))
print('Female:', str(len(test_female['MSE']['ADE_05'])))
```

#### Mean 

```python
mean_data = {'male': test_male['MEAN'], 'female': test_female['MEAN']}
pd.DataFrame.from_dict(mean_data, orient='index')
```

#### SD

```python
SD_data = {'male': test_male['SD'], 'female': test_female['SD']}
pd.DataFrame.from_dict(SD_data, orient='index')
```

## Box Plots - Test Split

#### ADE(0.5)

```python
fig, ax = plt.subplots(figsize = (8, 8))
data = [test_male['MSE']['ADE_05'], test_female['MSE']['ADE_05']]
ax.boxplot(data);
ax.set_title('ADE(0.5) With Outliers');
ax.set_ylabel('MSE');
plt.xticks([1, 2], ['male', 'female']);

fig, ax = plt.subplots(figsize = (8, 8))
ax.boxplot(data, showfliers = False);
ax.set_title('ADE(0.5) No Outliers');
ax.set_ylabel('MSE');
plt.xticks([1, 2], ['male', 'female']);
```
