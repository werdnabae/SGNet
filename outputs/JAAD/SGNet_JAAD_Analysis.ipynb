```python
import os
import numpy as np
import pickle as pkl
from matplotlib import pyplot as plt
import pandas as pd
```

# Data Split: Test

```python
test_adult = open('SGNet_JAAD_test_adult.pkl', 'rb')
test_adult = pkl.load(test_adult)
test_child = open('SGNet_JAAD_test_child.pkl', 'rb')
test_child = pkl.load(test_child)
test_elderly = open('SGNet_JAAD_test_elderly.pkl', 'rb')
test_elderly = pkl.load(test_elderly)
test_nolabel = open('SGNet_JAAD_test_no_label.pkl', 'rb')
test_nolabel = pkl.load(test_nolabel)
test_all = open('SGNet_JAAD_test_all.pkl', 'rb')
test_all = pkl.load(test_all)
```

## Number of Trajectories

#### Number of trajectories

```python
print('Child:', str(len(test_child['MSE']['ADE_05'])))
print('Adult:', str(len(test_adult['MSE']['ADE_05'])))
print('Elderly:', str(len(test_elderly['MSE']['ADE_05'])))
print('No label:', str(len(test_nolabel['MSE']['ADE_05'])))
print('All:', str(len(test_all['MSE']['ADE_05'])))
```

#### Mean 

```python
mean_data = {'child': test_child['MEAN'], 'adult': test_adult['MEAN'], 'elderly': test_elderly['MEAN'], 'no label': test_nolabel['MEAN'], 'all': test_all['MEAN']}
data = pd.DataFrame.from_dict(mean_data, orient='index')
data.round(decimals = 1)
```

#### SD

```python
SD_data = {'child': test_child['SD'], 'adult': test_adult['SD'], 'elderly': test_elderly['SD'], 'no label': test_nolabel['SD'], 'all': test_all['SD']}
pd.DataFrame.from_dict(SD_data, orient='index')
```

## Example Histogram

```python
fig, ax = plt.subplots(figsize = (10, 10))
count, edges, bars = ax.hist(test_child['MSE']['ADE_05'], bins = 50);
ax.set_xlabel('MSE')
ax.set_ylabel('# of trajectories')
plt.bar_label(bars);
```
