```python
import os
import numpy as np
import pickle as pkl
from matplotlib import pyplot as plt
import pandas as pd
```

# Data Split: Test

```python
test_male = open('SGNet_PIE_test__male_all.pkl', 'rb')
test_male = pkl.load(test_male)
test_female = open('SGNet_PIE_test__female_all.pkl', 'rb')
test_female = pkl.load(test_female)
test_all = open('SGNet_PIE_test_all.pkl', 'rb')
test_all = pkl.load(test_all)
```

## Statistics

#### Number of trajectories

```python
print('Male:', str(len(test_male['MSE']['ADE_05'])))
print('Female:', str(len(test_female['MSE']['ADE_05'])))
print('All:', str(len(test_all['MSE']['ADE_05'])))
```

#### Mean 

```python
mean_data = {'male': test_male['MEAN'], 'female': test_female['MEAN'], 'all': test_all['MEAN']}
pd.DataFrame.from_dict(mean_data, orient='index')
```

#### SD

```python
SD_data = {'male': test_male['SD'], 'female': test_female['SD'], 'all': test_all['SD']}
pd.DataFrame.from_dict(SD_data, orient='index')
```

## Example Histogram

```python
fig, ax = plt.subplots(figsize = (10, 10))
count, edges, bars = ax.hist(test_male['MSE']['ADE_05'], bins = 50);
ax.set_xlabel('MSE')
ax.set_ylabel('# of trajectories')
plt.bar_label(bars);
```

```python
fig, ax = plt.subplots(figsize = (10, 10))
count, edges, bars = ax.hist(test_female['MSE']['ADE_05'], bins = 50);
ax.set_xlabel('MSE')
ax.set_ylabel('# of trajectories')
plt.bar_label(bars);
```
